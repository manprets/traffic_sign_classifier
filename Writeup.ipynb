#### Step 3: Build Model

```python
# 1. Preprocess
#          Images in dataset are all resized to 32 x 32
#          Images were normalized using below logic:
#          X_train = (X_train-128.0)/256.0
#          The above transformation ensured that loss function would converge
#          All the channels (3 colors) in the images were used in the project 
#          All channels were chosen based on past performance
#          The reasoning is that color information helps deep learning network to infer 
#             labels from the images more accurately as we use all information present in 
#             the input images. Using graycale means losing information
# 2. Augument Dataset
#          There are 21 labels with number of images less than 500
#          These labels are good candidate for data augmentation
#          No data augumentation was performed in this project
# 3. Model Architecture
#          I used an architecture that I have used for SVHN dataset
#          The architecture is appropriate here as the type of problem is somewhat same:
#              Image Classification
#          Below is a breakdown of the architecture
#              a. Image Inpout (32x32x3)
#              b. Conv2d Layer (kernel=5, stride=1)
#              c. Relu Layer
#              d. MaxPool Layer (stride=2)
#              e. Conv2d Layer (kernel=5, stride=1)
#              f. Relu Layer
#              g. MaxPool Layer (stride=2)
#              h. Relu Layer
#              i. Dropout Layer
#              j. Fully Connected Layer
# 3. Model Hyper Parameters
#          Loss function: Cross Entropy
#          Optimizer: Adagrad
#          Learning rate: Exponential Decay
#          Batch Size: 64
# 4. Training Process
#          A few iteration were performed for training the model
#          The first iteration did not include regularization
#          In, the second iteration I increased number of features in every layer
#          For the third iteraion, I decreased the number of features at every layer
#             and introduced regularization to achieve satisfactory results
#          The third architecture was chosen as validation accuracy was > 0.93
# 5. Testing
#          The final model's test accuracy was: 0.943
#
```
